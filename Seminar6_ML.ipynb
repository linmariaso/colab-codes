```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn import datasets
```

```python
iris = datasets.load_iris()
iris
```

```python
data = pd.DataFrame(data = np.c_[iris['data'],iris['target']], columns=iris['feature_names']+['target'])
data
```

```python
setosa = data[data.species == 'setosa']
versicolor = data[data.species == 'versicolor']
virginica = data[data.species == 'virginica']

fig, ax = plt.subplots(figsize=(15,10))

ax.scatter(setosa['petal length (cm)'], setosa['petal width (cm)'], c='r', label='setosa')
ax.scatter(versicolor['petal length (cm)'], versicolor['petal width (cm)'], c='b', label='versicolor')
ax.scatter(virginica['petal length (cm)'], virginica['petal width (cm)'], c='g', label='virginica')

ax.set_xlabel('petal length (cm)')
ax.set_ylabel('petal width (cm)')
ax.legend()
ax.grid()
ax.set_title('Iris petals')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
from re import X
X = data.drop(['target','species'], axis=1)
X = X.to_numpy()[:, (2,3)]
y = data['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state = 5)
```

```python
from math import log
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
training = logreg.predict(X_train)
training
```

```python
test = logreg.predict(X_test)
test
```

```python
from sklearn import metrics

print(metrics.confusion_matrix(y_train, training))
print(metrics.classification_report(y_train, training))
```

```python
print(metrics.confusion_matrix(y_test, test))
print(metrics.classification_report(y_test, test))
```
