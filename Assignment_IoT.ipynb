```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
X = filtered[["soilmiosture", "temperature", "pressure", "altitude"]]
y = filtered["class"]
```

```python
scaler = StandardScaler()
X = scaler.fit_transform(X)
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

Random forest

```python
model = RandomForestClassifier(random_state=10)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
plt.figure(figsize=(10, 7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
plt.show()
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.tree import plot_tree
```

```python
individual_tree = model.estimators_[0]
```

```python
plt.figure(figsize=(12, 8))
feature_names = ["soilmiosture", "temperature", "pressure", "altitude"]
class_names = ['Very Dry', 'Dry','Wet','Very Wet']
plot_tree(individual_tree, feature_names=feature_names, class_names=class_names, filled=True)
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.inspection import DecisionBoundaryDisplay

clf = Pipeline(
    steps=[("scaler", StandardScaler()), ("knn", KNeighborsClassifier(n_neighbors=11))]
)
```

```python
_, axs = plt.subplots(ncols=2, figsize=(12, 5))

# Select two features for visualization (e.g., 'soilmiosture' and 'temperature')
features_to_visualize = ['soilmiosture', 'temperature']

for ax, weights in zip(axs, ("uniform", "distance")):
    clf.set_params(knn__weights=weights).fit(X_train[features_to_visualize], y_train)
    disp = DecisionBoundaryDisplay.from_estimator(
        clf,
        X_test[features_to_visualize],  # Use the selected features for visualization
        response_method="predict",
        plot_method="pcolormesh",
        xlabel=features_to_visualize[0],  # Update xlabel
        ylabel=features_to_visualize[1],  # Update ylabel
        shading="auto",
        alpha=0.5,
        ax=ax,
    )
    scatter = disp.ax_.scatter(X[features_to_visualize[0]], X[features_to_visualize[1]], c=y, edgecolors="k")  # Use selected features for scatter plot
    disp.ax_.legend(
        scatter.legend_elements()[0],
        class_names,
        loc="lower left",
        title="Classes",
    )
    _ = disp.ax_.set_title(
        f"3-Class classification\n(k={clf[-1].n_neighbors}, weights={weights!r})"
    )

plt.show()
```
