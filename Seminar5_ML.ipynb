4. Complete the ML stage in the Tutorial  using the Titanic dataset ("train.csv")

```python
data = pd.read_csv('train.csv')
```

```python
data.describe()
```

```python
data.head()
```

```python
data.isnull().sum()
```

```python
columns = ['Name', 'Ticket', 'Cabin']
```

```python
df = data.drop(columns, axis=1 )
df
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
new_def = []
cols = ['Pclass','Sex','Embarked']

for x in cols:
    new_def.append(pd.get_dummies(df[x]))

dummy = pd.concat(new_def, axis=1)
dummy
```

```python
df = pd.concat((df,dummy), axis=1)
df
```

```python
df.info()
```

```python
df['Age'] = df['Age'].interpolate()
```

```python
df.info()
```

```python
#change from pandas to numpy
X = df.values
y = df['Survived'].values

X = np.delete(X, 1, axis=1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```
